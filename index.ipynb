# PCA (Principal Component Analysis)

PCA (Principal Component Analysis) is a statistical technique used for dimensionality reduction while preserving as much variability (information) as possible in the data

### Concept:

    Goal: Reduce the number of variables in your dataset (i.e., lower the dimensionality), while retaining the important patterns.

    How: PCA transforms the original features into a new set of uncorrelated variables called principal components.

        These components are linear combinations of the original features.

        They are ranked so that the first component captures the most variance, the second the second most, and so on.

### When to use PCA

    To visualize high-dimensional data (e.g., plotting in 2D or 3D).

    To remove noise and redundancy.

    To speed up machine learning algorithms by reducing feature space.

## Steps in PCA:

    Standardize the data (mean = 0, variance = 1).

    Compute the covariance matrix.

    Calculate eigenvectors and eigenvalues of this matrix.

    Sort eigenvectors by descending eigenvalues (importance).

    Select top k components and project the data.

# imports

````python
import numpy as np
import seaborn as sns 
import pandas as pd 
from sklearn.datasets import make_classification
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt

````

````python
# generate sample classification data
X, y = make_classification(n_samples=100, n_features=20, n_informative=10, n_redundant=5, n_clusters_per_class=2, random_state=42)
````

````python
# stardardize the data
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
````

### With Specified Components 

````python
# Apply PCA to reduce the dimensionality of the data (with 2 components)
pca_5 = PCA(n_components=2)
X_pca_5 = pca_5.fit_transform(X_scaled)
````

````python
# Step 4: Plot the results
plt.figure(figsize=(8, 6))
plt.scatter(X_pca_5[:, 0], X_pca_5[:, 1], c=y, cmap='viridis')
plt.xlabel("Principal Component 1")
plt.ylabel("Principal Component 2")
plt.title("PCA - Reduced to 2D")
plt.grid(True)
plt.show()

# Step 5: Check explained variance
print("Explained variance ratio:", pca_5.explained_variance_ratio_)
````

### With specified Variance 

````python
# Pca_0.8 = PCA(n_components=0.8) 80 % variance
pca_08 = PCA(n_components=0.8)
pca_08 = pca_08.fit_transform(X_scaled)
````
